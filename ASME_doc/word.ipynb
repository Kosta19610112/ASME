```python
from docx import Document
import os
from PIL import Image
from io import BytesIO

def extract_docx_content(file_path):
    # Открываем документ
    doc = Document(file_path)
    
    # Извлекаем текст
    text = []
    for paragraph in doc.paragraphs:
        ttt = paragraph.text
        text.append(paragraph.text)
    
    # Извлекаем таблицы
    tables = []
    for table in doc.tables:
        table_data = []
        for row in table.rows:
            row_data = [cell.text for cell in row.cells]
            table_data.append(row_data)
        tables.append(table_data)
    
    # Извлекаем изображения
    images = []
    rels = doc.part.rels
    for rel in rels.values():
        if "image" in rel.target_ref:
            img_data = rel.target_part.blob
            images.append(Image.open(BytesIO(img_data)))
    
    return text, tables, images



# Пример использования
docx_file = "DOC/pdf_urez59_0.docx"
docx_file = "DOC/pdf_urez1_500.docx"
docx_file = "DOC/pdf_urez500_1000.docx"
docx_file = "DOC/pdf_urez1000_1496.docx"
docx_file = "DOC/ASME Sec II-D-Metric.docx"

text, tables, images = extract_docx_content(docx_file)

# Печать текста
#print("Текст документа:")
#print("\n".join(text))

# Печать таблиц
filename = 'Tab.txt'
file = open(filename, 'w')

for i, table in enumerate(tables):
    print(f"Таблица {i+1}:", file = file)
    for row in table:
#        print(row[1:], file = file)
        print(row, file = file)
file.close()

print('End Program')
# Сохранение изображений
#output_dir = "extracted_images"
#os.makedirs(output_dir, exist_ok=True)
#for i, img in enumerate(images):
#    img.save(os.path.join(output_dir, f"image_{i+1}.png"))
#    print(f"Сохранено изображение: image_{i+1}.png")
```

```python
from docx import Document

def get_line_from_word(file_path, line_number):
    """
    Читает строку с заданным номером из документа Word.
    :param file_path: Путь к файлу Word (.docx)
    :param line_number: Номер строки (начиная с 1)
    :return: Текст строки или сообщение об ошибке
    """
    try:
        doc = Document(file_path)
        lines = []
        # Собираем текст из всех абзацев
        for paragraph in doc.paragraphs:
            lines.extend(paragraph.text.splitlines())
        
        # Проверяем, существует ли строка с заданным номером
        if 0 < line_number <= len(lines):
            return lines[line_number - 1]
        else:
            return f"Строки с номером {line_number} нет в документе."
    except Exception as e:
        return f"Ошибка при чтении файла: {e}"

# Пример использования

docx_file = "DOC/pdf_urez59_0.docx"
#docx_file = "DOC/pdf_urez1_500.docx"
#docx_file = "DOC/pdf_urez500_1000.docx"
#docx_file = "DOC/pdf_urez1000_1496.docx"
#docx_file = "DOC/ASME Sec II-D-Metric.docx"

file_path = docx_file  # Укажите путь к файлу

for line_number in range(20):  # Номер строки, которую нужно прочитать
    result = get_line_from_word(file_path, line_number)
    print(result)
```

```python
import ast

def read_lines_from_file(filename, start, end):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            lines = file.readlines()
            return lines[start-1:end]  # Номера строк приводятся к индексации с 0
    except FileNotFoundError:
        print(f"Файл {filename} не найден.")
        return []
    except IndexError:
        print("Некорректные значения start или end.")
        return []


def find_line_with_text(filename, search_text):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            for i, line in enumerate(file, start=1):  # Нумерация строк с 1
                if search_text in line:
                    return i
        return None  # Если текст не найден
    except FileNotFoundError:
        print(f"Файл {filename} не найден.")
        return None

def stroka_v_spisok(input_str):
    # Преобразование строки в список
    try:
        result_list = ast.literal_eval(input_str)
    #    print(result_list)
    except (ValueError, SyntaxError):
        print("Ошибка: Некорректный формат строки.")
    return result_list

# Пример использования

"""def find_nomer_po_tekstu(filename, search_text):
    line_number = find_line_with_text(filename, search_text)
    if line_number:
        print(f"Текст '{search_text}' найден на строке {line_number}.")
    else:
        print(f"Текст '{search_text}' не найден в файле.")
    return line_number"""


filename = 'Tab.txt'
#search_text = "Таблица 18:"
search_text = "Таблица 19:"
No = find_line_with_text(filename, search_text)
No_header = No + 1

start = No_header  # Начальная строка (включительно)
end = No_header + 45   # Конечная строка (включительно)

Spisok = []
lines = read_lines_from_file(filename, start, end)
for line in lines:
    spisok = stroka_v_spisok(line)
    Spisok.append(spisok)

print(Spisok)

df = pd.DataFrame(Spisok[1:], columns=Spisok[0])


```
